```python
f_class_Age=df.groupby('Pclass')['Age'].mean()
f_class_Age=pd.DataFrame(f_class_Age)
f_class_Age
```

```python
a=list(f_class_Age['Age'])

def impute_age(cols):
  Age=cols[0]
  Pclass=cols[1]

  if pd.isnull(Age):

    if Pclass==1:
      return a[0]

    elif Pclass==2:
      return a[1]

    else:
      return a[2]
  else:
    return Age
```

```python
df['Age']=df[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
d=df.describe()
d=d.T
d.plot.bar(y="count")
plt.title("Bar plot of count of numeric features",fontsize=15)
```
